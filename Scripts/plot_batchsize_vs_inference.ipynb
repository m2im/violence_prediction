```python
import matplotlib.pyplot as plt
```

```python
# Data
batch_sizes = [1024, 512, 256, 128, 64, 32, 16, 8, 1]
inference_times = [9.85, 10.57, 10.65, 8.97, 8.95, 9.3, 10.15, 16.4, 75.52]  # Times in minutes
raw_times = ["9m51s", "10m34s", "10m39s", "8m58s", "8m57s", "9m18s", "10m09s", "16m24s", "1h15m31s"]
```

```python
# Create a figure and axis explicitly
fig, ax = plt.subplots(figsize=(10, 6))

# Scatter plot
ax.scatter(batch_sizes, inference_times, color='blue', s=100, label='Inference Time')

# Add labels to each point
for x, y in zip(batch_sizes, inference_times):
    ax.text(x, y + 1, f"{y:.2f} min", fontsize=10, ha='center', va='bottom')

# Customize axis labels and title
ax.set_xscale('log')  # Use logarithmic scale for proper spacing
ax.set_xticks(batch_sizes)
ax.set_xticklabels([str(b) for b in batch_sizes])  # Replace log values with batch sizes
ax.set_xlabel('Batch Size', fontsize=12)
ax.set_ylabel('Inference Time (minutes)', fontsize=12)
ax.set_title('Batch Size vs Inference Time', fontsize=14)
ax.grid(True, linestyle="--", linewidth=0.5)
# ax.legend()
```

```python
# Save the figure as EPS
plt.tight_layout()
fig.savefig("figs/batch_vs_inference.eps", format="eps", dpi=300, bbox_inches="tight")
```

```python
# Show plot
plt.show()
```
