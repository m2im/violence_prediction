```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Specify the paths to the .png files
image_paths = ["figs/ROC_Capitol_Hill.png", "figs/ROC_Israel.png", "figs/ROC_Russia_Ukraine.png", "figs/ROC_test_set.png"]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(6, 6))
```

```python
# Create a 2x2 subplot
fig, axes = plt.subplots(2, 2, figsize=(6, 6))

# Load and display each image in the subplot
for i, ax in enumerate(axes.flat):
    # Read the image
    img = mpimg.imread(image_paths[i])
    
    # Display the image
    ax.imshow(img)
    
    # Remove axis ticks
    ax.axis('off')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Save to disk
fig.savefig("figs/comparison_ROC_curves.eps", format="eps", dpi=300, bbox_inches="tight")
```
